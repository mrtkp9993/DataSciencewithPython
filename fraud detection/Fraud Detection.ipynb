```python
# Import required libraries
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, cohen_kappa_score
import pandas as pd
```

```python
# Import and quick look to data
df = pd.read_csv("creditcard.csv", sep=",")
df.head()
```

```python
# Delete time column - it's unnecessary to us
del df["Time"]
df.head()
```

```python
# First 29 columns are our X data and last column is our labels (y data)
X = df.ix[:, 0:29]
y = df.ix[:, 29]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Create a Random Forest Classifier
randomForestClassifier = RandomForestClassifier(n_estimators=10, n_jobs=-1) # n_jobs=-1 says use all cores on CPU
```

```python
# Fit RFClassifier
randomForestClassifier.fit(X_train, y_train)
```

```python
# Make prediction
y_pred = randomForestClassifier.predict(X_test)
```

```python
# Confusion matrix
cnf_matrix = confusion_matrix(y_pred, y_test)
print(cnf_matrix)
```

```python
# Accuracy, Recall and Cohen's Cappa
accuracy = np.round(100*float((cnf_matrix[0][0]+cnf_matrix[1][1]))/float((cnf_matrix[0][0]+cnf_matrix[1][1] + cnf_matrix[1][0] + cnf_matrix[0][1])),2)
recall = np.round(100*float((cnf_matrix[1][1]))/float((cnf_matrix[1][0]+cnf_matrix[1][1])),2)
cappa = np.round(cohen_kappa_score(y_test, y_pred),3)
print(accuracy, recall, cappa)
```
